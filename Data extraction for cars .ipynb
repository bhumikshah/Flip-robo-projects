```python
import selenium
from selenium import webdriver
import time
```

```python
driver1=webdriver.Chrome(r"C:\Users\Bhumik Shah\Downloads\chromedriver_win32 (5)\chromedriver.exe")
```

```python
# looking for used cars in Thane locations and extrating url of used cars
url='https://www.cardekho.com/used-cars+in+thane'
```

```python
driver1.get(url)
```

```python
from selenium.common.exceptions import NoSuchElementException
cars_details=[]
used_kilometer=[]
owner=[]
variant=[]
mileage=[]
fuel=[]
location=[]
price=[]
# Extracting out xpath for car details with date of manufacturing, brand and model details
try:
    details=driver1.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']//a")
    for i in details:
        cars_details.append(i.text)
except NoSuchElementException:
    cars_details.append('-')
# Extracting out the xpath for variant details whether it is automatic or manual
try:
    det1=driver1.find_elements_by_xpath("//div[@class='truncate dotlist']//span[3]")
    for j in det1:
        variant.append(j.text)
except NoSuchElementException:
    variant.append('-')
#Extracting out the xpath for kilometer runned by hte cars
try:
    det2=driver1.find_elements_by_xpath("//div[@class='truncate dotlist']//span[1]")
    for k in det2:
            used_kilometer.append(k.text)
except NoSuchElementException:
    used_kilometer.append('-')
# Extracting out the xpath for the fuel details
try:
    det3=driver1.find_elements_by_xpath("//div[@class='truncate dotlist']//span[2]")
    for l in det3:
        fuel.append(l.text)
except NoSuchElementException:
    fuel.append('-')
try:
    det4=driver1.find_elements_by_xpath("//div[@class='cityIcon']")
    for m in det4:
        location.append(m.text)
except NoSuchElementException:
    location.append('-')
# Extracting out the xpath for car price
try:
    det5=driver1.find_elements_by_xpath("//span[@class='amnt ']")
    for n in det5:
        price.append(n.text)
except NoSuchElementException:
    price.append('-')
```

```python
len(location)
```

```python
urls2=[]
# Extracting out the urls of all the used cars in Thane to extract out the owner details from inside of each url
details=driver1.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']//a")
for i in details:
    urls2.append(i.get_attribute('href'))
```

```python
len(urls2)
```

```python
for b in urls2[:364]:
    driver1.get(b)
    try:
        det6=driver1.find_element_by_xpath("//li[@class='gsc_col-xs-6'][4]//div//div")
        owner.append(det6.text)
    except NoSuchElementException:
        owner.append('-')
    try:
        det7=driver1.find_element_by_xpath("//ul[@class='gsc_row']//li[1]//div[2]")
        mileage.append(det7.text)
    except NoSuchElementException:
        mileage.append('-')
```

```python
mileage
```

```python
import pandas as pd
df=pd.DataFrame({})
df['car_details']=cars_details
df['variant']=variant
df['used_kilometer']=used_kilometer
df['fuel']=fuel
df['owner']=owner
df['location']=location
df['mileage']=mileage
df['Price']=price
```

```python
df
```

```python
df.to_csv('cars details.csv')
```

```python
cars_details7=[]
used_kilometer7=[]
owner7=[]
variant7=[]
fuel7=[]
price7=[]
# Extracting out xpath for car details with date of manufacturing, brand and model details
try:
    details=driver8.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']//a[1]")
    for i in details:
        cars_details7.append(i.text)
except NoSuchElementException:
    cars_details7.append('-')
# Extracting out the xpath for variant details whether it is automatic or manual
try:
    det1=driver8.find_elements_by_xpath("//div[@class='truncate dotlist']//span[3]")
    for j in det1:
        variant7.append(j.text)
except NoSuchElementException:
    variant7.append('-')
#Extracting out the xpath for kilometer runned by hte cars
try:
    det2=driver8.find_elements_by_xpath("//div[@class='truncate dotlist']//span[1]")
    for k in det2:
            used_kilometer7.append(k.text)
except NoSuchElementException:
    used_kilometer7.append('-')
# Extracting out the xpath for the fuel details
try:
    det3=driver8.find_elements_by_xpath("//div[@class='truncate dotlist']//span[2]")
    for l in det3:
        fuel7.append(l.text)
except NoSuchElementException:
    fuel7.append('-')
# Extracting out the xpath for car price
try:
    det5=driver8.find_elements_by_xpath("//span[@class='amnt ']")
    for n in det5:
        price7.append(n.text)
except NoSuchElementException:
    price7.append('-')
```

```python
len(price7)
```
