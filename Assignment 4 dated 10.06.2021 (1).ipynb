```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/'
```

```python
driver.get(url)
```

```python
# Extracting top 30 most viewed videos on youtube from wikipedia
from selenium.common.exceptions import NoSuchElementException
searchbar=driver.find_element_by_id('searchInput')
searchbar.send_keys('most viewed youtube videos')
search=driver.find_element_by_id('searchButton')
search.click()
rank=[]
Video_name=[]
artist=[]
views=[]
upload_date=[]
for i in range(1,31):
# Extracting rank details from x path
    try:
        d1=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr[{}]//td[1]".format(i))
        rank.append(d1.text)
    except NoSuchElementException:
        d2=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")
        rank.append(d2.text)
# Extracting video name from x path
    try:
        d3=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr[{}]//td[2]".format(i))
        Video_name.append(d3.text)
    except NoSuchElementException:
        d4=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")
        Video_name.append(d4.text)
# Extracting artist details from xpath
    try:
        d5=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr[{}]//td[3]".format(i))
        artist.append(d5.text)
    except NoSuchElementException:
        d6=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")
        artist.append(d6.text)
#Extracting views from xpath
    try:
        d7=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr[{}]//td[4]".format(i))
        views.append(d7.text)
    except NoSuchElementException:
        d8=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")
        views.append(d8.text)
# Extracting upload_date from xpath
    try:
        d9=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr[{}]//td[5]".format(i))
        upload_date.append(d9.text)
    except NoSuchElementException:
        d10=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")
        upload_date.append(d10.text)
```

```python
df=pd.DataFrame({})
df['rank']=rank
df['name']=Video_name
df['artist']=artist
df['views in billions']=views
df['upload_date']=upload_date
```
