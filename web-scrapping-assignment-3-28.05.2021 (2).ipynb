```python
# scrape all product urls from pages
from selenium.common.exceptions import NoSuchElementException
import time
product_urls = []
start=0
end=3
for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')#scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))#appending the urls in product_urls list
    nxt_button=driver.find_element_by_xpath("//li[@class='a-last']//a")#scraping the list of buttons from the page
    nxt_button.click()
    time.sleep(2)

#Make empty list
Brand = []
Description=[]
price=[]
reviews=[]
noofreviews=[]
returns=[]
Delivery=[]
Availability=[]
otherdetails=[]


for url in product_urls:
    driver.get(url)
    time.sleep(2)
    # Extracting Brand from xpath
    try:
        brand = driver.find_element_by_xpath('//a[@id="bylineInfo"]')      
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')  
    #Extracting description from xpath
    try:
        describe = driver.find_element_by_xpath('//span[@id="productTitle"]')      
        Description.append(describe.text)
    except NoSuchElementException:
        Description.append('-') 
     # Extracting price from xoath   
    try:
        pri=driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        price.append(pri.text)
    except NoSuchElementException:
        price.append('-')
      # Extracting reviews from xpath  
    try:
        rev=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        reviews.append(rev.text)
    except NoSuchElementException:
        reviews.append('-')
    # Extracting global reviews from xpath
    try:
        nor=driver.find_element_by_xpath("//span[@class='a-size-base a-color-secondary']")
        noofreviews.append(nor.text)
    except NoSuchElementException:
        noofreviews.append('-')
     # Extracting Returns/Exchange from xpath   
    try:
        ret=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        returns.append(ret.text)
    except NoSuchElementException:
        returns.append('-')
    # Extracting Delivery date from xpath
    try:
        ed=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        Delivery.append(ed.text)
    except NoSuchElementException:
        Delivery.append('-')
     # Extracting Availability from xpath   
    try:
        ava=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
        Availability.append(ava.text)
    except:
        Availability.append('-')
    # Extracting other details from xpath
    try:
        oth=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']//span")
        otherdetails.append(oth.text)
    except NoSuchElementException:
        otherdetails.append('-')            
```

```python
df=pd.DataFrame({})
df['Brand']=Brand
df['Description']=Description
df['Price']=price
df['reviews']=reviews
df['Product Url']=product_urls
df['no of reviews']=noofreviews
df['returns']=returns
df['Delivery']=Delivery
df['Availability']=Availability
df['other details']=otherdetails
```

```python
df1=pd.DataFrame({})
df1['Date']=Date
df1['Startup Name']=Startup_Name
df1['Industry/Vertical']=Industry_Vertical
df1['Sub Vertical']=Sub_Vertical
df1['City/Location']=City_Location
df1['Investors Name']=Investors_Name
df1['Investment Type']=Investment_Type
df1['Amount (in USD)']=Amount_in_USD
```
