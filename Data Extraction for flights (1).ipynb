```python
#  looking for flights available from New Delhi to Mumbai on easymytrip.com for the journey date 05.10.2021
url='https://flight.easemytrip.com/FlightList/Index?srch=DEL-Delhi-India|BOM-Mumbai-India|05/10/2021&px=1-0-0&cbn=0&ar=undefined&isow=true&isdm=true&lng=&CouponCode=&ompAff=&bc=&ISWL='
driver1.get(url)
```

```python
from selenium.common.exceptions import NoSuchElementException
Airline_name=[]
source=[]
Destination=[]
departure_time=[]
arrival_time=[]
duration=[]
total_stops=[]
price=[]
# Extracting X path for the airline name
try:
    det=driver1.find_elements_by_xpath("//span[@class='txt-r4 ng-binding']")
    for i in det:
        Airline_name.append(i.text)
except NoSuchElementException:
    Airline_name.append('-')
# Extracting xpath for source i.e. New Delhi in your case
try:
    det1=driver1.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']//span[2]")
    for j in det1:
        source.append(j.text)
except NoSuchElementException:
    source.append('-')
# Extracting xpath for destination i.e. Mumbai in your case
try:
    det2=driver1.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']//span[2]")
    for k in det2:
        Destination.append(k.text)
except NoSuchElementException:
    Destination.append('-')
# Extracting xpath for the departure time from the source i.e New delhi in our case
try:
    det3=driver1.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']//span[1]")
    for l in det3:
        departure_time.append(l.text)
except NoSuchElementException:
    departure_time.append('-')
# Extracting xpath for the arrival time at the destination i.e Mumbai in our case
try:
    det4=driver1.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[1]")
    for m in det4:
        arrival_time.append(m.text)
except NoSuchElementException:
    arrival_time.append('-')
# Extracting xpath for the duration of the journey
try:
    det5=driver1.find_elements_by_xpath("//span[@class='dura_md ng-binding']")
    for n in det5:
        duration.append(n.text)
except NoSuchElementException:
    duration.append('-')
# Extracting xpath for the total stops during the journey from source to destination
try:
    det6=driver1.find_elements_by_xpath("//span[@class='dura_md2 ng-scope']")
    for o in det6:
        total_stops.append(o.text)
except NoSuchElementException:
    total_stops.append('-')
# Extracting xpath for the price per ticket of the journey
try:
    det7=driver1.find_elements_by_xpath("//div[@class='col-md-8 col-sm-8 col-xs-9 txt-r6-n ng-binding']")
    for b in det7:
        price.append(b.text)
except NoSuchElementException:
    price.append('-')
```

```python
journey_date=[]
i=0
while i<171:
    a= '05/10/2021'
    journey_date.append(a)
    i+=1
```

```python
import pandas as pd
df=pd.DataFrame({})
df['journey_date']=journey_date
df['Airline_name']=Airline_name
df['source']=source
df['Destination']=Destination
df['departure_time']=departure_time
df['arrival_time']=arrival_time
df['Duration']=duration
df['total_stops']=total_stops
df['Price']=price
```

```python
df.to_csv('05.10.2021.csv')
```

```python
from selenium.common.exceptions import NoSuchElementException
Airline_name6=[]
source6=[]
Destination6=[]
departure_time6=[]
arrival_time6=[]
duration6=[]
total_stops6=[]
price6=[]
# Extracting X path for the airline name
try:
    det=driver7.find_elements_by_xpath("//span[@class='txt-r4 ng-binding']")
    for i in det:
        Airline_name6.append(i.text)
except NoSuchElementException:
    Airline_name6.append('-')
# Extracting xpath for source i.e. New Delhi in your case
try:
    det1=driver7.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']//span[2]")
    for j in det1:
        source6.append(j.text)
except NoSuchElementException:
    source6.append('-')
# Extracting xpath for destination i.e. Mumbai in your case
try:
    det2=driver7.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']//span[2]")
    for k in det2:
        Destination6.append(k.text)
except NoSuchElementException:
    Destination6.append('-')
# Extracting xpath for the departure time from the source i.e New delhi in our case
try:
    det3=driver7.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']//span[1]")
    for l in det3:
        departure_time6.append(l.text)
except NoSuchElementException:
    departure_time6.append('-')
# Extracting xpath for the arrival time at the destination i.e Mumbai in our case
try:
    det4=driver7.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[1]")
    for m in det4:
        arrival_time6.append(m.text)
except NoSuchElementException:
    arrival_time6.append('-')
# Extracting xpath for the duration of the journey
try:
    det5=driver7.find_elements_by_xpath("//span[@class='dura_md ng-binding']")
    for n in det5:
        duration6.append(n.text)
except NoSuchElementException:
    duration6.append('-')
# Extracting xpath for the total stops during the journey from source to destination
try:
    det6=driver7.find_elements_by_xpath("//span[@class='dura_md2 ng-scope']")
    for o in det6:
        total_stops6.append(o.text)
except NoSuchElementException:
    total_stops6.append('-')
# Extracting xpath for the price per ticket of the journey
try:
    det7=driver7.find_elements_by_xpath("//div[@class='col-md-8 col-sm-8 col-xs-9 txt-r6-n ng-binding']")
    for b in det7:
        price6.append(b.text)
except NoSuchElementException:
    price6.append('-')
```

```python
journey_date6=[]
i=0
while i<217:
    a= '11/10/2021'
    journey_date6.append(a)
    i+=1
```

```python
import pandas as pd
df6=pd.DataFrame({})
df6['journey_date']=journey_date6
df6['Airline_name']=Airline_name6
df6['source']=source6
df6['Destination']=Destination6
df6['departure_time']=departure_time6
df6['arrival_time']=arrival_time6
df6['Duration']=duration6
df6['total_stops']=total_stops6
df6['Price']=price6
```
